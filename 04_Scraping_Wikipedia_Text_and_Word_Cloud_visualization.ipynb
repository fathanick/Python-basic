```
import wikipedia
import re
```

```
#specify the title of Wikipedia page 
wiki = wikipedia.page('Universitas Islam Indonesia')

#Extract the text from the page
text = wiki.content

# Clean text
text = re.sub(r'==.*?==+', '', text)
text = text.replace('\n', '')
print(text)
```

```
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```
#add parameters to word cloud
wcld = WordCloud(width=500,
                 height=300,
                 max_font_size=100,
                 min_font_size=10,
                 max_words=100,
                 background_color='black',)
wcld.generate(text)

#plot/show image
plt.imshow(wcld, interpolation='bilinear') # .imshow will display data as an image
plt.axis("off") #hide axis values
plt.show() #show data
```
